```python
## 3.Use z-score normalization to standardize the values of the income 
## attribute. [Do not change the original income attribute in the table.]
```

```python
bank_data['income_znorm'] = (bank_data['income'] - bank_data['income'].mean())/bank_data['income'].std()
```

```python
## 4.Discretize the age attribute into 3 categories (corresponding to "young", 
## "mid-age", and "old"). [Do not change the original age attribute in the
## table.]
```

```python
## Relatively Arbitrarily binning into young, mid age, and old
## Note this is based on my perception of age groupings!

bank_data['age_bins'] = pd.cut(bank_data['age'], [0, 35, 60, 100], labels=['young', 'mid', 'old'])
```

```python
## 5. Use Min-Max Normalization to transform the values of all numeric attributes
## (income, age, children) in the original table (before the transformations
## in parts 3 and 4 above) onto the range 0.0-1.0.
```

```python
## NOTE: I LEFT THE ORIGINAL AGE,INCOME VALUES TO USE IN #8
## WHICH REQUESTED A SCATTER PLOT BASED ON NON-NORMALIZED DATA

min_inc = bank_data['income'].min()
max_inc = bank_data['income'].max()
bank_data['income_norm'] = [(x-min_inc)/(max_inc-min_inc) for x in bank_data['income']]

min_age = bank_data['age'].min()
max_age = bank_data['age'].max()
bank_data['age_norm'] = [(x-min_age)/(max_age-min_age) for x in bank_data['age']]

min_kid = bank_data['children'].min()
max_kid = bank_data['children'].max()
bank_data['children_norm'] = [(x-min_kid)/(max_kid-min_kid) for x in bank_data['children']]

bank_data
```

```python
## 6.Convert the table (after normalization in part 5) into the standard 
## spreadsheet format. Note that this
## requires converting each categorical attribute into multiple binary ("dummy")
## attributes (one for each values
## of the categorical attribute) and assigning binary values corresponding to the
## presence or not presence of
## the attribute value in the original record). The numeric attributes should
## remain unchanged. Save this new
## table into a file called bank_numeric.csv and submit it along with your 
## assignment. [Hint: you might
## consider using the get_dummies for Pandas data frames.]
```

```python
## NOTE: When creating dummy variables, since a lot of the values are simply
## 'yes/no', when looking at the sheet you get a bunch of YES NO YES NO YES NO
## attribute labels, which is very confusing... so, I relabeled the yes/no's
## to give a hint as to what they're being used for so they'd be more
## descriptive.


mask = (bank_data.married == 'YES')
column_name = 'married'
bank_data.loc[mask, column_name] = 'MARRIED_YES'

mask = (bank_data.married == 'NO')
column_name = 'married'
bank_data.loc[mask, column_name] = 'MARRIED_NO'

mask = (bank_data.car == 'YES')
column_name = 'car'
bank_data.loc[mask, column_name] = 'CAR_YES'

mask = (bank_data.car == 'NO')
column_name = 'car'
bank_data.loc[mask, column_name] = 'CAR_NO'

mask = (bank_data.savings_acct == 'YES')
column_name = 'savings_acct'
bank_data.loc[mask, column_name] = 'SAVINGS_YES'

mask = (bank_data.savings_acct == 'NO')
column_name = 'savings_acct'
bank_data.loc[mask, column_name] = 'SAVINGS_NO'

mask = (bank_data.current_acct == 'YES')
column_name = 'current_acct'
bank_data.loc[mask, column_name] = 'CURRENT_YES'

mask = (bank_data.current_acct == 'NO')
column_name = 'current_acct'
bank_data.loc[mask, column_name] = 'CURRENT_NO'

mask = (bank_data.mortgage == 'YES')
column_name = 'mortgage'
bank_data.loc[mask, column_name] = 'MORTGAGE_YES'

mask = (bank_data.mortgage == 'NO')
column_name = 'mortgage'
bank_data.loc[mask, column_name] = 'MORTGAGE_NO'

mask = (bank_data.pep == 'YES')
column_name = 'pep'
bank_data.loc[mask, column_name] = 'PEP_YES'

mask = (bank_data.pep == 'NO')
column_name = 'pep'
bank_data.loc[mask, column_name] = 'PEP_NO'
```

```python
##Dummy Variable setup (manual)
dummy_gender = pd.get_dummies(bank_data['gender'])
bank_data = pd.concat([bank_data, dummy_gender], axis = 1)

dummy_region = pd.get_dummies(bank_data['region'])
bank_data = pd.concat([bank_data, dummy_region], axis = 1)

dummy_married = pd.get_dummies(bank_data['married'])
bank_data = pd.concat([bank_data, dummy_married], axis = 1)

dummy_car = pd.get_dummies(bank_data['car'])
bank_data = pd.concat([bank_data, dummy_car], axis = 1)

dummy_savings = pd.get_dummies(bank_data['savings_acct'])
bank_data = pd.concat([bank_data, dummy_savings], axis = 1)

dummy_current = pd.get_dummies(bank_data['current_acct'])
bank_data = pd.concat([bank_data, dummy_current], axis = 1)

dummy_mortgage = pd.get_dummies(bank_data['mortgage'])
bank_data = pd.concat([bank_data, dummy_mortgage], axis = 1)

dummy_pep = pd.get_dummies(bank_data['pep'])
bank_data = pd.concat([bank_data, dummy_pep], axis = 1)

dummy_age_cats = pd.get_dummies(bank_data['age_bins'])
bank_data = pd.concat([bank_data, dummy_age_cats], axis = 1)
```

```python
## Dropping categorical data
bank_data.drop(['gender', 'region', 'married', 'car', 'savings_acct', 'current_acct', 'mortgage', 'pep', 'age_bins'], axis=1)
```

```python
## Saving to CSV
bank_data.to_csv('bank_numeric.csv', sep='\t')
```

```python
N = 4
pep_yes = (34, 46, 71, 123)
pep_no = (28, 50, 102, 146)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, pep_yes, width, color='r')
rects2 = ax.bar(ind + width, pep_no, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('PEP by Region')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Inner City', 'Town', 'Rural', 'Suburban'))

ax.legend((rects1[0], rects2[0]), ('Yes PEP', 'No PEP'))


def autolabel(rects):

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
